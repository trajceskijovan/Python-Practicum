### Question 4

Make a data set called *dallas_sub* that includes data only from the city of `Dallas` in 2012 & 2013.

```python
dallas_sub=txhousing.query('year == 2012 & city == "Dallas" | year==2013 & city=="Dallas"')
dallas_sub
```

```python
dallas_sub = txhousing[(txhousing['city'] == "Dallas") & (txhousing['year'].isin([2012, 2013]))]
dallas_sub
```

### Question 5

Add a column to the *dallas_sub* data set called *prct_sold* that calculates the percentage of _listings_ that were sold (sales/listings * 100). Be sure to **save** the results also as a data frame called *dallas_sub*.

```python
# https://www.geeksforgeeks.org/adding-new-column-to-existing-dataframe-in-pandas/
dallas_sub = dallas_sub.assign(prct_sold=dallas_sub["sales"]/dallas_sub["listings"]*100)
dallas_sub.head()
```

### Question 6

Calculate the **average** percentage of _listings_ that were sold in Dallas **in each month of the year** based on your *dallas_sub* data set. Save the results of the calculation in an data frame called *dallas_summary*.

```python
# name new df
# group by month
# present AVG of listings
dallas_summary = dallas_sub.groupby(['month'])\
.agg({"prct_sold": "mean"})

#.agg({"prct_sold": ["mean","std"]})
# https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.core.groupby.DataFrameGroupBy.agg.html
dallas_summary
```

### Question 7

Arrange the *dallas_summary* in **descending** order based on the average percentage of listings that were sold in Dallas in 2012, so you can see **which month** had the greatest percentage of houses sold in Dallas on average from 2012-2013. You do not need to save the results.

```python
# max to min
dallas_summary.reset_index()\
    .sort_values(["prct_sold"], ascending=False)
```

### Question 9

In 2012, in which month were the most houses sold in Texas? 

```python
# can use sum too

txhousing[txhousing['year'] == 2012]\
    .groupby(['month'])\
    .agg({"sales": "max"})\
    .reset_index()\
    .sort_values(['sales'], ascending=False)
```
